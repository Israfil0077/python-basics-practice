## Graph

```python
import heapq
import math

graph = {
    'A': [('B', 4), ('C', 3)],
    'B': [('A', 4), ('D', 5), ('E', 12)],
    'C': [('A', 3), ('D', 7), ('F', 10)],
    'D': [('B', 5), ('C', 7), ('E', 2)],
    'E': [('D', 2), ('B', 12), ('G', 3)],
    'F': [('C', 10), ('G', 5)],
    'G': [('E', 3), ('F', 5)]
}

coords = {
    'A': (1,6),
    'B': (3,7),
    'C': (2,4),
    'D': (5,5),
    'E': (7,5),
    'F': (4,2),
    'G': (8,3)
}

def heuristic(a, b):
    x1, y1 = coords[a]
    x2, y2 = coords[b]

    return math.sqrt((x1-x2)**2 + (y1-y2)**2)


def shortest_path(graph, start, goal, mode):

    pq = []
    heapq.heappush(pq, (0, start))

    dist = {start: 0}
    parent = {}   # for path reconstruction
    visited = 0

    while pq:
        visited += 1
        priority, node = heapq.heappop(pq)

        if node == goal:
            break

        for neighbor, weight in graph[node]:
            new_cost = dist[node] + weight

            if neighbor not in dist or new_cost < dist[neighbor]:
                dist[neighbor] = new_cost
                parent[neighbor] = node

            if mode == 'dijkstra':
                priority = new_cost
            elif mode == 'astar':
                priority = new_cost + heuristic(neighbor, goal)

        heapq.heappush(pq, (priority, neighbor))

    path = []
    node = goal

    while node != start:
        path.append(node)
        node = parent[node]

    path.append(start)
    path.reverse()

    return visited, path, dist[goal]


if __name__ == "__main__":
    start = "A"
    goal = "G"

    # Run Dijkstra
    vis1, path1, cost1 = shortest_path(graph, start, goal, mode="dijkstra")

    print("Dijkstra:")
    print("Visited:", vis1)
    print("Path:", path1)
    print("Cost:", cost1)

    print()

    # Run A*
    vis2, path2, cost2 = shortest_path(graph, start, goal, mode="astar")

    print("A*:")
    print("Visited:", vis2)
    print("Path:", path2)
    print("Cost:", cost2)
```
